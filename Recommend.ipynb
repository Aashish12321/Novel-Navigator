```python
# It pushes those users(only user_id) to the overlap_users who likes the same books as in liked_books.
overlap_users = set()

with open("Ratings.csv", 'r') as f:
    for line in f:
        try:
            user_id, book_id, rating = line.strip().split(",")
            # do something with user_id, book_id, and rating
        except ValueError:
            # skip the line if it doesn't contain three comma-separated values
            continue
    
        if user_id in overlap_users:
            continue

        if book_id in liked_books and int(rating) >= 4:
                overlap_users.add(user_id)
```

```python
# overlap_users
len(list(overlap_users))
```

```python
# It pushes users(along with user_id, book_id, rating) with all the books they have rated, into the rec_lines, who likes the liked_books.
rec_lines = []

with open("Ratings.csv", 'r') as f:
    for line in f:
        try:
            user_id, book_id, rating = line.strip().split(",")
        except ValueError:
            continue
        
        if user_id in overlap_users:
            rec_lines.append([user_id, book_id, rating])
```

```python
# rec_lines
len(rec_lines)
```

```python
rec_lines[0:3]
```

```python
import pandas as pd

recs = pd.DataFrame(rec_lines, columns=["user_id", "ISBN", "rating"])
recs["ISBN"] = recs["ISBN"].astype(str)
```

```python
# Users with all the books they have rated, who likes the liked_books.
recs
```

```python
# Books rated by the most users from recs is stored in top_recs.
top_recs = recs["ISBN"].value_counts().head(5)
```

```python
# Top 5 books similar to liked_books with most users. 
top_recs
```

```python
top_recs = top_recs.index.values
```

```python
books = pd.read_csv("Books.csv")
ratings = pd.read_csv("Ratings.csv")
books_titles = ratings.merge(books,on='ISBN')
books_titles['ISBN'] = books_titles['ISBN'].astype(str)
```

```python
all_recs = recs["ISBN"].value_counts()
```

```python
all_recs = all_recs.to_frame().reset_index()
all_recs.columns = ["ISBN", "book_count"]
```

```python
all_recs.head(5)
```

```python
all_recs = all_recs.merge(books_titles, how="inner", on="ISBN")
all_recs
```

```python
all_recs["score"] = all_recs["book_count"] * (all_recs["book_count"] / all_recs["Book-Rating"])
```

```python
all_recs.sort_values("score", ascending=False).head(10)
```
